```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

mnist = datasets.load_digits()
X = mnist['data']
y = mnist['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=(6/7),
                                                    random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import time


forest_clf = RandomForestClassifier(random_state=42)
start = time.time()
forest_clf.fit(X_train, y_train)
end = time.time()
print('training time:', end-start)
print(accuracy_score(y_test, forest_clf.predict(X_test)))
```

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=0.95)
X_train_transformed = pca.fit_transform(X_train)
X_test_transformed = pca.transform(X_test)

start = time.time()
forest_clf.fit(X_train_transformed, y_train)
end = time.time()

print('training time:', end-start)
print(accuracy_score(y_test, forest_clf.predict(X_test_transformed)))
```
